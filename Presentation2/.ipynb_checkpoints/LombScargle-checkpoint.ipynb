```python
import matplotlib.pyplot as plt
import numpy as np

from scipy import signal
%matplotlib inline
```

```python
def lombscargle(t, x, f):
    x -= np.mean(x)
    
    omegas = 2 * np.pi * f
    taus = (0.5 / omegas) * np.arctan(
        np.sum(np.sin(2 * omegas * t))
        / np.sum(np.cos(2 * omegas * t))
    )
    
    pgram = np.zeros(f.shape)
    for indx, (omega, tau) in enumerate(zip(omegas, taus)):
        sines = np.sin(omega * (t - tau))
        cosines = np.cos(omega * (t - tau))
        pgram[indx] = 0.5 * (
            np.sum(x * cosines)**2 / np.sum(cosines**2)
            + np.sum(x * sines)**2 / np.sum(sines**2)
        )
    
    return pgram
```

# Periodograms

The periodogram is a tool to estimate the power spectral density of a signal. This characterizes the signal in terms of the power of each frequency component.

The power spectral density of a continuous function $x(t)$ is simply the square of its Fourier transform
$$ S(\omega) = \left\| \mathscr{F} \left\{ x(t) \right\} \right\|^2 = \left\| X(\omega) \right\|^2 $$

This can also be applied to the discrete Fourier transform
$$ \begin{aligned}
S(\omega) &= \frac{1}{N} \left\| \sum_{j=1}^{N} x(t_j) e^{-i \omega t_j} \right\|^2 \\
          &= \frac{1}{N} \left( \left( \sum_{j=1}^{N} x_j \cos(\omega t_j) \right)^2 + \left( \sum_{j=1}^{N} x_j \sin(\omega t_j) \right)^2 \right)
   \end{aligned} $$
This is the form of the classical periodogram.

There are two main reasons why the classical periodogram is not used anymore:
- Statisitical Problems: the function $S(\omega)$ is very noisy, even if the data is not, and does not decrease when the sample size is increased. This means that the spectrum doesn't converge, no matter the sample size or time interval chosen.
- Spectral Leakage: for a sinusoidal signal with frequency $\omega_0$, the power in the periodogram leaks to other frequencies other than $\omega_0$, particularly leakage from higher frequencies to much lower frequencies.

```python
t = np.linspace(0, 2 * np.pi, 400)
x = 0.5 * np.sin(2 * np.pi * 10 * t) + 0.8 * np.sin(2 * np.pi * 25 * t) + 0.2 * np.random.normal(size=400)

plt.scatter(t, x, s=0.5)
```

```python
freq = np.fft.fftfreq(t.size, t[1] - t[0])
pgram = 1 / x.size * np.abs(np.fft.fft(x))**2

plt.plot(freq, pgram)
plt.xlim(5, 30)
```

```python
freq, pgram = signal.periodogram(x, 1 / (t[1] - t[0]))
plt.plot(freq, pgram)
plt.xlim(5, 30)
```

```python
freq = np.linspace(5, 30, x.size)
pgram = lombscargle(t, x, freq)
plt.plot(freq, pgram)
```

# Least-Squares Spectal Analysis

So what do we do if the signal cannot be uniformly sampled?
- This often happens when the measurements are weather dependent, such as for astronomy

The frequency spectrum can be estimated by taking the least-squares fit of sinusoids to the data samples.

This method was first proposed by Petr Venicek, and is sometimes called the Vanicek method.

The frequencies are chosen, usually uniformly over the range of frequencies of interest, then optimized by weighting them so the residuals are minimized.
- The data vector $\vec{\phi}$ is approximated by a weighted sum of sinusoidal functions in the matrix $A$ with the weight vector $\vec{w}$
$$ \vec{\phi} \approx A \vec{w}$$
- The weight vector is chosen to minimize the sum of squared errors when $\vec{\phi}$ is approximated
$$ \sigma = \sum_{n=1}^{N} (\phi_n - x_n)^2 $$
- The closed solution for $\vec{w}$ expressed as
$$ \vec{w} = (A^T A)^{-1} A^T \vec{\phi} $$
- When the basis functions in $A$ are orthogonal, the matrix $A^T A$ is a diagonal matrix and makes the inversion process easy. This is why Scargle proposed these simplifications, and created the constraint
$$ \sum_{n=1}^N \cos(\omega t_n - \tau) \sin(\omega t_n - \tau) = 0$$
where $\tau$ is a small time delay that forces the pair of sinusoids to be mutually orthogonal at sample times $t_n$. This constraint is equivalent to
$$ \tan(2 \omega \tau) = \frac{\sum_{n=1}^N \sin(2 \omega t_n)}{\sum_{n=1}^N \cos(2 \omega t_n)}$$

```python
t = np.linspace(0, 10 * np.pi, 1000).tolist()
t = t[0:50] + t[100:250] + t[350:700] + t[800:900]
t = np.array(t)
x = 0.5 * np.sin(2 * np.pi * 10 * t) + 0.8 * np.sin(2 * np.pi * 25 * t) + 0.2 * np.random.normal(t.size)

plt.scatter(t, x, s=0.5)
```

```python
freq = np.linspace(4, 30, x.size)
ls_pgram = signal.lombscargle(t, x, 2 * np.pi * freq, precenter=True)

plt.plot(freq, ls_pgram)
```

```python
plt.plot(freq, lombscargle(t, x, freq))
```
