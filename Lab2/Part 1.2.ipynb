```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

t = np.linspace(0.0, 1.0, 1024)
freq = np.fft.fftfreq(t.size, 1.0 / 1024.0)

# Two distinct sinusoids
f1 = np.sin(2 * np.pi * 100 * t) + 0.5 * np.sin(2 * np.pi * 200 * t)

# Slight difference from f1
f2 = np.sin(2 * np.pi * 100.5 * t) + 0.5 * np.sin(2 * np.pi * 200 * t)

# Amplitude modulation
f3 = (2 + np.sin(2 * np.pi * 8 * t)) * np.sin(2 * np.pi * 100 * t)

# Frequency modulation
f4 = np.sin(2 * np.pi * 100 * (1 + 0.1 * np.sin(2 * np.pi * 8 * t)) * t)

# Plotting
fig, subs = plt.subplots(nrows=2, ncols=4, figsize=(13, 5), dpi=300)

subs[0, 0].set_title('Two Sinusoids')
subs[0, 0].set_xlim(-250, +250)
subs[0, 0].set_ylabel('Amplitude')
subs[0, 0].plot(freq, np.abs(np.fft.fft(f1)) * 2 / t.size)
subs[1, 0].set_xlim(-250, +250)
subs[1, 0].set_xlabel('Frequency (Hz)')
subs[1, 0].set_ylabel('Power ($\\times 10^{-1}$)')
subs[1, 0].plot(freq, np.abs(np.fft.fft(f1) * 2 / t.size)**2 * 10)

subs[0, 1].set_title('Slightly Changed Sinusoids')
subs[0, 1].set_xlim(-250, +250)
subs[0, 1].set_ylim(0, 1)
subs[0, 1].plot(freq, np.abs(np.fft.fft(f2)) * 2 / t.size)
subs[1, 1].set_xlim(-250, +250)
subs[1, 1].set_ylim(0, 10)
subs[1, 1].set_xlabel('Frequency (Hz)')
subs[1, 1].plot(freq, np.abs(np.fft.fft(f2) * 2 / t.size)**2 * 10)

subs[0, 2].set_title('Amplitude Modulation')
subs[0, 2].set_xlim(-150, +150)
subs[0, 2].plot(freq, np.abs(np.fft.fft(f3)) * 2 / t.size)
subs[1, 2].set_xlim(-150, +150)
subs[1, 2].set_xlabel('Frequency (Hz)')
subs[1, 2].plot(freq, np.abs(np.fft.fft(f3) * 2 / t.size)**2 * 10)

subs[0, 3].set_title('Frequency Modulation')
subs[1, 3].set_xlabel('Frequency (Hz)')
subs[0, 3].plot(freq, np.abs(np.fft.fft(f4)) * 2 / t.size)
subs[1, 3].plot(freq, np.abs(np.fft.fft(f4) * 2 / t.size)**2 * 10)

plt.savefig('fft_examples')
```
