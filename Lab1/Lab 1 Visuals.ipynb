```python
#Photon Graphs
txt_data = open('histograms.txt', 'r')
tau = []
theta = []
phi = []

for indx in range(0,1000000):
    line = txt_data.readline()
    line = line[:-1]
    tau.append(float(line[:17].replace(" ","")))
    theta.append(float(line[17:34].replace(" ","")))
    phi.append(float(line[34:].replace(" ","")))
    
tau = np.asarray(tau)
theta = np.asarray(theta)
phi = np.asarray(phi)
    
plt.figure(0)
n, bins, s = plt.hist(tau, bins=20)
x = np.linspace(0,10,99)
y = np.exp(-1*x)*np.max(n)
plt.plot(x,y)
plt.xlabel(r'$\tau$')
plt.ylabel(r'$P(\tau)$')
plt.savefig('fig6.png')

plt.figure(1)
n, b, s = plt.hist(theta,bins=20)
x = np.linspace(0,np.pi,99)
y = np.sin(x)*np.max(n)
plt.plot(x,y)
plt.xlabel(r'$\theta$')
plt.ylabel(r'$P(\theta)$')
plt.savefig('fig7.png')

plt.figure(2)
n, b, s = plt.hist(phi,bins=10)
x= np.linspace(0,np.pi*2,99)
y= np.ones(99)*np.max(n)
plt.plot(x,y)
plt.xlabel(r'$\phi$')
plt.ylabel(r'$P(\phi)$')
plt.savefig('fig8.png')
```

```python
#Scattering Intensity
txt_data = open('photon_output.txt', 'r')
mu = []

for indx in range(0,50273):
    line = txt_data.readline()
    line = line[:-1]
    mu.append(float(line.replace(" ","")))

plt.figure(0)
n, bins, patches = plt.hist(mu, bins=20)
mu_coord = np.linspace(0.025,0.975,20)

I = np.zeros(20)
for indx, Ei in enumerate(n):
    I[indx] = (Ei*20)/(2*50273*mu_coord[indx])
    
print(n)
    
Iexpected = 0.0244 * ( 51.6 - 0.0043 * (np.arccos(mu_coord) * 180 / np.pi)**2)
    
plt.figure(1)
plt.plot(mu_coord, I)
plt.plot(mu_coord, Iexpected)
```

```python
#Markov Chain Graphs

txt_data = open('markovdata.txt', 'r')
n = []
p_sun = []
p_notsun = []
p_notrain = []
p_rain = []

for indx in range(0,100):
    line = txt_data.readline()
    line = line[:-1]
    n.append(int(line[:12]))
    p_sun.append(float(line[12:25]))
    p_notsun.append(float(line[25:42]))
    p_notrain.append(float(line[42:59]))
    p_rain.append(float(line[59:]))

plt.figure(0)
f, ax = plt.subplots(2,1, sharex=True)
ax[0].plot(n, p_sun, 'r', n, p_notsun, 'b')
ax[0].set_ylabel('Probability')
ax[1].plot(n, p_notrain, 'r', n, p_rain, 'b')
ax[1].set_ylabel('Probability')
ax[1].set_xlabel('Number of days (n)')
f.savefig('fig9.png')

txt_data.readline()
txt_data.readline()
txt_data.readline()

n = np.zeros(31)
p = np.zeros((12,31))

for indx in range(0,31):
    line = txt_data.readline()
    line = line[:-1]
    n[indx] = (int(line[:12]))
    p[0,indx] =(float(line[12:25]))
             
for indx in range(1,11):
    txt_data.readline()
    txt_data.readline()
    txt_data.readline()
    for indx2 in range(0,31):
        line = txt_data.readline()
        line = line[:-1]
        p[indx, indx2] =(float(line[12:25]))
             
plt.figure(2)
for i in range(0,11):
    plt.plot(n,p[i,:], label=str(i*0.1))
plt.xlabel('Number of Days (n)')
plt.ylabel('Probability (n)')
plt.legend(loc='lower right', shadow=True, fontsize='small')
plt.savefig('fig10.png')
```

```python
#Metropolis

txt_data = open('metropolis_data.txt', 'r')
i = np.zeros(1000)
values = np.zeros(1000)
accepted = np.zeros(1000)

for indx in range(0,1000):
    line = txt_data.readline()
    line = line[:-1]
    i[indx] = int(line[:12])
    values[indx] = float(line[12:29])
    accepted[indx] = float(line[29:])
    
acceptance_rate = np.mean(accepted)
print(acceptance_rate)

#Theoretical values
x_vals = np.linspace(-1.5, 2., 99)
y = (np.sin(5*x_vals) + np.sin(2*x_vals) +2) * np.exp(-(x_vals**2))

plt.figure(0)
n, bins, passes = plt.hist(values,bins=20)
print(n)
plt.plot(x_vals, y * (159/3.256))

for indx in range(0,1000):
    line = txt_data.readline()
    line = line[:-1]
    i[indx] = int(line[:12])
    values[indx] = float(line[12:29])
    accepted[indx] = float(line[29:])
    
acceptance_rate = np.mean(accepted)
print(acceptance_rate)

plt.figure(1)
n, bins, passes = plt.hist(values,bins=20)
print(n)
plt.plot(x_vals, y*(160/3.256))

for indx in range(0,1000):
    line = txt_data.readline()
    line = line[:-1]
    i[indx] = int(line[:12])
    values[indx] = float(line[12:29])
    accepted[indx] = float(line[29:])
    
acceptance_rate = np.mean(accepted)
print(acceptance_rate)

plt.figure(2)
n, bins, passes = plt.hist(values,bins=20)
print(n)
plt.plot(x_vals, y*(165/3.256))
```
